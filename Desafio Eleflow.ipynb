# C) Calcule a variância e a variação absoluta do máximo, mínimo de cada cidade, mês a mês.

```python
# Colunas solicitadas transformadas de string para float
df["PREÇO MÍNIMO REVENDA"] = df["PREÇO MÍNIMO REVENDA"].str.replace(",",".")
df["PREÇO MÍNIMO REVENDA"] = df["PREÇO MÍNIMO REVENDA"].astype(float)
df["PREÇO MÁXIMO REVENDA"] = df["PREÇO MÁXIMO REVENDA"].str.replace(",",".")
df["PREÇO MÁXIMO REVENDA"] = df["PREÇO MÁXIMO REVENDA"].astype(float)
df.dtypes
```

> **Outro método de agrupação de dados além de pivot_table do pandas é o groupby**

```python
df_new = df.groupby(["mês", "REGIÃO", "ESTADO", "MUNICÍPIO"]).mean()
```

```python
df_new["VARIAÇÃO ABSOLUTA"] = df_new["PREÇO MÁXIMO REVENDA"] - df_new["PREÇO MÍNIMO REVENDA"]
df_new
```

>**Variância do Máximo e Mínimo**

```python
np.var(df_new["PREÇO MÁXIMO REVENDA"], axis=0)
```

```python
np.var(df_new["PREÇO MÍNIMO REVENDA"], axis=0)
```

# D) Quais são as 5 cidades que possuem a maior diferença entre o combustível mais barato e o mais caro

```python
# Analisando mensalmente, o top 5 de maior diferença é:
df_new_2 = df_new.sort_values("VARIAÇÃO ABSOLUTA", ascending = [False])
df_new_2.head(5)
```
