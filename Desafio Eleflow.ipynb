# A) Estes valores estão distribuídos em dados semanais, agrupe eles por mês e calcule as médias de valores de cada combustível por cidade.

### Criar coluna datetime para a criação de coluna mês

```python
df["datafinal_dt"] =  pd.to_datetime(df['DATA FINAL'], format='%d/%m/%Y')
```

```python
df["mês"] = df["datafinal_dt"].dt.month
## Verifica se a criação da coluna mês ocorreu corretamente
df.head(20000)
```

```python
# inspeciona os tipos dos dados da database
df.dtypes
```

```python
# Torna a coluna de valor em float

df["PREÇO MÉDIO REVENDA"] = df['PREÇO MÉDIO REVENDA'].str.replace(",",".")
df["PREÇO MÉDIO REVENDA"] = df['PREÇO MÉDIO REVENDA'].astype(float)
```

```python
df.dtypes
```

## Agregação e cálculo do preço médio de revenda municipal pelo Pandas
